```python
import numpy as np
import pandas as pd
import colorsys
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
dataset_url = "https://raw.githubusercontent.com/bahl24/pH-data-prediction/master/ph-data.csv"
df = pd.read_csv(dataset_url)

df.head()
```

```python
print("Columns:", df.columns)
```

```python
# Normalize RGB to 0-1
rgb = df[['red', 'green', 'blue']].values / 255.0
# Vectorized conversion
hsv = np.array([colorsys.rgb_to_hsv(r, g, b) for r, g, b in rgb])
```

```python
#adding HSV columns

df['hue'] = hsv[:, 0]
df['saturation'] = hsv[:, 1]
df['value'] = hsv[:, 2]

df[['hue', 'saturation', 'value']].head()
```

```python
#independent and dependentvariables
X = df[['hue', 'saturation', 'value']]
y = df['label']
```

```python
#polynomial feature
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_poly, y, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#evaluating the model
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (pH units):", mae)
```
